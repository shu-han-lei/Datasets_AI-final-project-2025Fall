HIGH-INCOME (120–150% AMI)


```python
def get_income_category(wage, year, df_HUD):
    thr30 = df_HUD.loc[df_HUD["Year"] == year, "30% AMI"].values[0]
    thr50 = df_HUD.loc[df_HUD["Year"] == year, "50% AMI"].values[0]
    thr60 = df_HUD.loc[df_HUD["Year"] == year, "60% AMI"].values[0]
    thr80 = df_HUD.loc[df_HUD["Year"] == year, "80% AMI"].values[0]
    thr100 = df_HUD.loc[df_HUD["Year"] == year, "100% AMI"].values[0]
    thr120 = df_HUD.loc[df_HUD["Year"] == year, "120% AMI"].values[0]

    # ---- LOW-INCOME ----
    if wage < thr30:
        return "Low-30"
    elif wage < thr50:
        return "Low-50"
    elif wage < thr60:
        return "Low-60"
    elif wage < thr80:
        return "Low-80"

    # ---- MEDIAN-INCOME ----
    elif wage < thr120:
        return "Median"

    # ---- HIGH-INCOME ----
    else:
        return "High"

```

### Yearly JAFS Pipeline -  OD × WAC Detailed Merged Dataset (with w_tract)

```python
base = "/content/drive/MyDrive/AI_DATA"
jafs_dir = Path(f"{base}/JAFS_BRONX_wage_HUD")
jafs_dir.mkdir(exist_ok=True)
```

```python
all_year_dfs = []
years_sorted = sorted(data_od.keys())   # ['2010', ..., '2022']

for year in years_sorted:
    print(f"\n=== Year {year} ===")
    year_int = int(year)

    # ---------- 1) 取当年 OD / WAC / RAC ----------
    df_od  = data_od[year]      # Bronx home OD
    df_wac = data_wac[year]     # MSA WAC
    df_rac = data_rac[year]     # Bronx-only RAC

    # WAC 只取需要列
    wac_use_cols = ["w_tract", "year", "C000"] + CNS_cols
    df_wac_use = df_wac[wac_use_cols].copy()

    # ---------- 2) OD + WAC merge ----------
    df = df_od.merge(df_wac_use, on=["w_tract", "year"], how="left")
    df[["C000"] + CNS_cols] = df[["C000"] + CNS_cols].fillna(0)

    # ---------- 3) 当年 HUD 阈值 ----------
    limits  = df_HUD.loc[year_int]
    thr30   = limits["30% AMI"]
    thr50   = limits["50% AMI"]
    thr60   = limits["60% AMI"]
    thr80   = limits["80% AMI"]
    thr120  = limits["120% AMI"]    # 用于 MEDIAN / HIGH 分界

    # ---------- 4) 当年各行业年收入（OES） ----------
    wages = wage_lookup[year_int]   # dict: CNSxx -> annual wage

    # 大类：LOW / MEDIAN / HIGH
    low_cns     = [c for c in CNS_cols if wages[c] <= thr80]
    median_cns  = [c for c in CNS_cols if thr80 < wages[c] <= thr120]
    high_cns    = [c for c in CNS_cols if wages[c] > thr120]

    # LOW 里面的细分：30 / 50 / 60 / 80
    low30_cns = [c for c in CNS_cols if wages[c] <= thr30]
    low50_cns = [c for c in CNS_cols if thr30 < wages[c] <= thr50]
    low60_cns = [c for c in CNS_cols if thr50 < wages[c] <= thr60]
    low80_cns = [c for c in CNS_cols if thr60 < wages[c] <= thr80]

    print("  low_cns   :", low_cns)
    print("  median_cns:", median_cns)
    print("  high_cns  :", high_cns)

    # ---------- 5) 计算各类可达岗位（flow S000 × jobs） ----------
    df["access_jobs_total"] = df["S000"] * df["C000"]

    # 大类
    df["access_low_total"]    = df["S000"] * df[low_cns].sum(axis=1)
    df["access_median_total"] = df["S000"] * df[median_cns].sum(axis=1)
    df["access_high_total"]   = df["S000"] * df[high_cns].sum(axis=1)

    # LOW 子类（可选，如果你只想大类，可以跳过）
    df["access_low30"] = df["S000"] * (df[low30_cns].sum(axis=1) if low30_cns else 0)
    df["access_low50"] = df["S000"] * (df[low50_cns].sum(axis=1) if low50_cns else 0)
    df["access_low60"] = df["S000"] * (df[low60_cns].sum(axis=1) if low60_cns else 0)
    df["access_low80"] = df["S000"] * (df[low80_cns].sum(axis=1) if low80_cns else 0)

    # ---------- 6) 聚合到 Bronx home tract ----------
    agg_cols = [
        "access_jobs_total",
        "access_low_total", "access_median_total", "access_high_total",
        "access_low30", "access_low50", "access_low60", "access_low80"
    ]

    df_jafs = (
        df.groupby(["h_tract", "year"])[agg_cols]
          .sum()
          .reset_index()
    )

    # ---------- 7) 合并 RAC（人口作为分母） ----------
    df_rac_use = (
        df_rac[["h_tract", "C000"]]
        .rename(columns={"C000": "resident_pop"})
        .drop_duplicates("h_tract")
    )

    df_jafs = df_jafs.merge(df_rac_use, on="h_tract", how="left")
    denom = df_jafs["resident_pop"].replace(0, pd.NA)

    # per capita 指标
    df_jafs["JAFS_total_pc"]   = df_jafs["access_jobs_total"]   / denom
    df_jafs["JAFS_low_pc"]     = df_jafs["access_low_total"]    / denom
    df_jafs["JAFS_median_pc"]  = df_jafs["access_median_total"] / denom
    df_jafs["JAFS_high_pc"]    = df_jafs["access_high_total"]   / denom

    df_jafs["JAFS_low30_pc"]   = df_jafs["access_low30"] / denom
    df_jafs["JAFS_low50_pc"]   = df_jafs["access_low50"] / denom
    df_jafs["JAFS_low60_pc"]   = df_jafs["access_low60"] / denom
    df_jafs["JAFS_low80_pc"]   = df_jafs["access_low80"] / denom

    # ---------- 8) 保存该年份结果 ----------
    out_path = jafs_dir / f"JAFS_bronx_HUDwage2_{year}.parquet"
    df_jafs.to_parquet(out_path, index=False)
    print(f"  Saved {out_path.name}, rows={len(df_jafs)}")

    all_year_dfs.append(df_jafs)

# ---------- 9) 拼成 panel ----------
jafs_panel = pd.concat(all_year_dfs, ignore_index=True)
panel_path = jafs_dir / "JAFS_bronx_HUDwage_panel_2010_2022.parquet"
jafs_panel.to_parquet(panel_path, index=False)
print(f"\nPanel dataset saved: {panel_path}")
```

```python
# ==== Macro Sector ====
macro_sector_map = {
    "Office": [
        "CNS09", "CNS10", "CNS11", "CNS12", "CNS13"
    ],
    "Industrial": [
        "CNS03", "CNS04", "CNS05", "CNS06", "CNS08"
    ],
    "Institutional": [
        "CNS15", "CNS16"
    ],
    "LocalServices": [
        "CNS07", "CNS14", "CNS17", "CNS18", "CNS19"
    ]
}
```

```python
all_year_dfs = []
years_sorted = sorted(data_od.keys())   # ['2010', ..., '2022']

for year in years_sorted:
    print(f"\n=== Year {year} ===")
    year_int = int(year)

    # ---------- 1) 取当年 OD / WAC / RAC ----------
    df_od  = data_od[year]      # Bronx home OD
    df_wac = data_wac[year]     # MSA WAC
    df_rac = data_rac[year]     # Bronx-only RAC

    # WAC 只取需要列
    wac_use_cols = ["w_tract", "year", "C000"] + CNS_cols
    df_wac_use = df_wac[wac_use_cols].copy()

    # ---------- 2) OD + WAC merge ----------
    df = df_od.merge(df_wac_use, on=["w_tract", "year"], how="left")
    df[["C000"] + CNS_cols] = df[["C000"] + CNS_cols].fillna(0)

    # ---------- 3) 当年 HUD 阈值 ----------
    limits  = df_HUD.loc[year_int]
    thr30   = limits["30% AMI"]
    thr50   = limits["50% AMI"]
    thr60   = limits["60% AMI"]
    thr80   = limits["80% AMI"]
    thr120  = limits["120% AMI"]    # 用于 MEDIAN / HIGH 分界

    # ---------- 4) 当年各行业年收入（OES） ----------
    wages = wage_lookup[year_int]   # dict: CNSxx -> annual wage

    # 大类：LOW / MEDIAN / HIGH
    low_cns     = [c for c in CNS_cols if wages[c] <= thr80]
    median_cns  = [c for c in CNS_cols if thr80 < wages[c] <= thr120]
    high_cns    = [c for c in CNS_cols if wages[c] > thr120]

    # LOW 里面的细分：30 / 50 / 60 / 80
    low30_cns = [c for c in CNS_cols if wages[c] <= thr30]
    low50_cns = [c for c in CNS_cols if thr30 < wages[c] <= thr50]
    low60_cns = [c for c in CNS_cols if thr50 < wages[c] <= thr60]
    low80_cns = [c for c in CNS_cols if thr60 < wages[c] <= thr80]

    print("  low_cns   :", low_cns)
    print("  median_cns:", median_cns)
    print("  high_cns  :", high_cns)

    # ---------- 5) 计算各类可达岗位（flow S000 × jobs） ----------
    df["access_jobs_total"] = df["S000"] * df["C000"]

    df["access_low_total"]    = df["S000"] * (df[low_cns].sum(axis=1) if low_cns else 0)
    df["access_median_total"] = df["S000"] * (df[median_cns].sum(axis=1) if median_cns else 0)
    df["access_high_total"]   = df["S000"] * (df[high_cns].sum(axis=1) if high_cns else 0)

    df["access_low30"] = df["S000"] * (df[low30_cns].sum(axis=1) if low30_cns else 0)
    df["access_low50"] = df["S000"] * (df[low50_cns].sum(axis=1) if low50_cns else 0)
    df["access_low60"] = df["S000"] * (df[low60_cns].sum(axis=1) if low60_cns else 0)
    df["access_low80"] = df["S000"] * (df[low80_cns].sum(axis=1) if low80_cns else 0)

    # ---------- 5c) Macro Sector job totals & accessibility ----------
    for sector, cns_list in macro_sector_map.items():
        valid_cns = [c for c in cns_list if c in df.columns]
        if valid_cns:
            df[f"jobs_{sector}"]    = df[valid_cns].sum(axis=1)
            df[f"access_{sector}"]  = df["S000"] * df[f"jobs_{sector}"]
        else:
            df[f"jobs_{sector}"]    = 0
            df[f"access_{sector}"]  = 0

    # ---------- 5b) 保存 OD × WAC 详细数据（含 w_tract + Macro Sector） ----------
    detail_cols = (
        ["h_tract", "w_tract", "year", "S000"] +
        ["C000"] + CNS_cols +
        ["access_jobs_total", "access_low_total",
         "access_median_total", "access_high_total"] +
        [f"jobs_{s}" for s in macro_sector_map.keys()] +
        [f"access_{s}" for s in macro_sector_map.keys()]
    )

    df_detail = df[detail_cols].copy()
    detail_path = jafs_dir / f"OD_WAC_detail_{year}.parquet"
    df_detail.to_parquet(detail_path, index=False)
    print(f"  Saved detailed OD-WAC dataset: {detail_path.name}, rows={len(df_detail)}")

    # ---------- 6) 聚合到 Bronx home tract ----------
    agg_cols = [
        "access_jobs_total",
        "access_low_total", "access_median_total", "access_high_total",
        "access_low30", "access_low50", "access_low60", "access_low80"
    ] + [f"access_{s}" for s in macro_sector_map.keys()]

    df_jafs = (
        df.groupby(["h_tract", "year"])[agg_cols]
          .sum()
          .reset_index()
    )

    # ---------- 7) 合并 RAC（人口作为分母） ----------
    df_rac_use = (
        df_rac[["h_tract", "C000"]]
        .rename(columns={"C000": "resident_pop"})
        .drop_duplicates("h_tract")
    )

    df_jafs = df_jafs.merge(df_rac_use, on="h_tract", how="left")
    denom = df_jafs["resident_pop"].replace(0, pd.NA)

    # per capita 指标（工资桶）
    df_jafs["JAFS_total_pc"]   = df_jafs["access_jobs_total"]   / denom
    df_jafs["JAFS_low_pc"]     = df_jafs["access_low_total"]    / denom
    df_jafs["JAFS_median_pc"]  = df_jafs["access_median_total"] / denom
    df_jafs["JAFS_high_pc"]    = df_jafs["access_high_total"]   / denom

    df_jafs["JAFS_low30_pc"]   = df_jafs["access_low30"] / denom
    df_jafs["JAFS_low50_pc"]   = df_jafs["access_low50"] / denom
    df_jafs["JAFS_low60_pc"]   = df_jafs["access_low60"] / denom
    df_jafs["JAFS_low80_pc"]   = df_jafs["access_low80"] / denom

    # per capita 指标（Macro Sectors）
    for sector in macro_sector_map.keys():
        df_jafs[f"JAFS_{sector}_pc"] = df_jafs[f"access_{sector}"] / denom

    # ---------- 8) 保存该年份结果 ----------
    out_path = jafs_dir / f"JAFS_bronx_HUDwage_macro_{year}.parquet"
    df_jafs.to_parquet(out_path, index=False)
    print(f"  Saved {out_path.name}, rows={len(df_jafs)}")

    all_year_dfs.append(df_jafs)

# ---------- 9) 拼成 panel ----------
jafs_panel = pd.concat(all_year_dfs, ignore_index=True)
panel_path = jafs_dir / "JAFS_bronx_HUDwage_macro_panel_2010_2022.parquet"
jafs_panel.to_parquet(panel_path, index=False)
print(f"\nPanel dataset saved: {panel_path}")
```
